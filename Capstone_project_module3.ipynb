Reading the data postal data from Wikipedia 

```python
df1=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

Creadting the DataFrame

```python
postal_df=df1[0]
```

Removing the Not assigned Borough rows 

```python
postal_df=postal_df[postal_df['Borough'] != 'Not assigned'].reset_index(drop=True)
```

Reading the latitude and longitude from the csv file

```python
geo_df=pd.read_csv('Geospatial_Coordinates.csv')
```

```python
geo_df.head()
```

Merging the postal code with geo spatial details and making one dataframe

```python
postal_geo_df=pd.merge(postal_df, geo_df,  how='inner')
```

```python
postal_geo_df.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(postal_geo_df['Borough'].unique()),
        postal_geo_df.shape[0]
    )
)
```

```python
postal_geo_df['Borough'].unique()
```

Getting the Latitude and langitude values for Toronto

```python
filter_list=['Downtown Toronto','East Toronto','West Toronto','Central Toronto']
toronto_df=postal_geo_df[postal_geo_df['Borough'].isin(filter_list)].reset_index(drop=True)
```

Toronato neighborhood list and its shape

```python
toronto_df.shape
```

```python
toronto_df
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Totonto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Totonto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Borough'], toronto_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
